```python
import pandas as pd
```

```python
# List of CSV files to concatenate
file_list = ['JC-202306-citibike-tripdata.csv', 'JC-202401-citibike-tripdata.csv']

# Create an empty list to store DataFrames
dfs = []

# Read each CSV file and append to the list of DataFrames
for file in file_list:
    dfs.append(pd.read_csv(file))

# Concatenate all DataFrames into one
concatenated_df = pd.concat(dfs, ignore_index=True)
```

```python
# Show the dataframe
concatenated_df
```

```python
# Check data types
concatenated_df.dtypes
```

```python
# Change data types and reduce decimal places
concatenated_df["started_at"] = pd.to_datetime(concatenated_df["started_at"])
concatenated_df["ended_at"] = pd.to_datetime(concatenated_df["ended_at"])
concatenated_df["start_lat"] = concatenated_df["start_lat"].apply(lambda x: round(x,4))  
concatenated_df["end_lat"] = concatenated_df["end_lat"].apply(lambda x: round(x,4))  
concatenated_df["start_lng"] = concatenated_df["start_lng"].apply(lambda x: round(x,4))  
concatenated_df["end_lng"] = concatenated_df["end_lng"].apply(lambda x: round(x,4))  
```

```python
concatenated_df
```

```python
# Check data types again
concatenated_df.dtypes
```

```python
# Export concatenated & cleaned DataFrame to single CSV file
concatenated_df.to_csv('cleaned_2023-2024.csv', index=False)
```
